```python
# Problem 1: Unique Words in a Sentence

sentence = "python is easy and python is powerful"

words = sentence.split()
# Convert list into set to remove duplicates
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
     
```

```python
#Problem 2: Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_paid = max(employees, key=employees.get)
print("Highest Salary:", highest_paid, "-", employees[highest_paid])
```

```python
#Problem 3: Find Maximum and Minimum Values

# Input list
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Output
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
#Problem 4: Count Products Above 1000

# Input list
prices = [450, 1200, 899, 1500, 300]

# Initialize counter
count = 0

# Loop through prices
for price in prices:
    if price > 1000:
        count += 1

# Output
print("Products above 1000:", count)
```

```python
#Problem  5: Calculate Attendance Percentage

# Input list
attendance = ["P", "P", "A", "P", "P"]

# Count total days
total_days = len(attendance)

# Count number of Present days
present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

# Calculate percentage
percentage = (present_count / total_days) * 100

# Output
print("Attendance Percentage:", percentage)
```

```python
# Problem 6: Remove Duplicate Phone Numbers
# Input list
phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

# Output
print("Unique phone numbers:", unique_numbers)
```

```python
# Problem 7: Count Character Frequency

# Input string
text = "python"

# Create empty dictionary
frequency = {}

# Loop through each character
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

# Output
print(frequency)
```

```python
# Problem 9: Check if Key Exists in Dictionary

employees_data = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

key_to_check = "Ravi"

if key_to_check in employees_data:
    print("Employee exists")
else:
    print("Employee does not exist")
```
